```python
import seaborn as sns
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

# Crear el diagrama de dispersión con personalización
sns.scatterplot(x=x, y=y, color='g', size=100)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Diagrama de dispersión')
plt.show()
```
